### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

```python
Matplotlib allowsfor plotting in either of two ways: pyplot or object-oriented approach. The recommended approach is the object-oriented approach for more control and customization.
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
1: Figure: The whole figure, which may contain one or more axes.
2: Axes: A subplot or a set of plots within a figure. It is the area where data is plotted
3: Axis: The number line or a 2D space where the data is plotted. It includes tick marks, labels, and the axis title.
4: Artist: Everything that is drawn on the figure is an artist, including figures, axes, and other elements.       
```

### When plotting in Matplotlib, what is the expected input data type?

```python
When plotting in Matplotlib, the expected input data type is typically a NumPy array or array-like data structures.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x_values = np.linspace(-4, 4, 100)
y_values = x_values ** 2

# Object-oriented approach
fig, ax = plt.subplots()
ax.plot(x_values, y_values, label='y = x^2')
ax.set_title('Plot of y = x^2')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend()
plt.show()

# Pyplot approach
plt.plot(x_values, y_values, label='y = x^2')
plt.title('Plot of y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x_scatter = np.linspace(-2, 2, 100)
y_scatter = x_scatter ** 2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# Creating subplots
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

# Scatter plot
ax1.scatter(random_data_x, random_data_y, alpha=0.5, label='Random Data')
ax1.plot(x_scatter, y_scatter, color='red', label='y = x^2')
ax1.set_title('Scatter Plot and Line Plot')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.legend()

# Bar plot
ax2.bar(names, values, color='green')
ax2.set_title('Bar Plot of Fruit Data')
ax2.set_xlabel('Fruits')
ax2.set_ylabel('Values')

# Adjust layout for better visualization
plt.tight_layout()
plt.show()
```
